```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter02/Dataset/Boston.csv")
```

```python
df
```

```python
df.dropna()
```

```python
df=df.dropna()
```

```python
df=df.drop_duplicates()
```

```python
newdf= df.rename(columns= {'CRIM':'crimeratepercapita','ZN':'landover25k_dqft','INDUS' :'non-retaillandproptn' , 'CHAS':'riverdummy', 'NOX':'nitricoxide',
                           'RM':'avgnoroomsperdwelling' , 'AGE':'proptnowneroccupied' , 'DIS':'weighteddist' , 'RAD':'radialhighwayaccess' , 'TAX':'proptaxrate', 'PTRATIO':'pupilteacherratio' , 'LSTAT':'pctlowerstatus' , 'MEDV':'medianvalue_ks'})
```

```python
newdf
```

```python
x= newdf.drop('crimeratepercapita', axis=1)
y=newdf[['crimeratepercapita']]
seed=10
test_data_size= 0.3
x_train ,x_test, y_train, y_test= train_test_split(x,y,test_size=test_data_size, random_state=seed)
```

```python
train_data= pd.concat([x_train,y_train] , axis=1)
test_data= pd.concat([x_test, y_test], axis=1)
```

```python
multilinearmodel = smf.ols(formula='np.log(crimeratepercapita) ~ (pctlowerstatus + radialhighwayaccess + medianvalue_ks + nitricoxide)**2' , data=train_data) 
```

```python
multiloglinear= multilinearmodel.fit()
```

```python
print(multiloglinear.summary())
```
